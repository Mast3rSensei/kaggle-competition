```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sales = pd.read_csv('train.csv')
```

```python
sales.head()
```

```python
sales.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
sales.dtypes
```

```python
sales.isna().sum()
```

```python
sales[(sales['Open'] == 0) & (sales['Sales'] > 0)]
```

```python
new_sales = sales.drop(np.where(sales['Open'] == 0)[0])
```

```python
new_sales['State_holiday'].value_counts()
```

```python
new_sales = pd.get_dummies(new_sales, columns=['State_holiday'], drop_first=True)
new_sales
```

```python
new_sales['Year'] = pd.DatetimeIndex(new_sales['Date']).year
new_sales['Month'] = pd.DatetimeIndex(new_sales['Date']).month
new_sales['Day'] = pd.DatetimeIndex(new_sales['Date']).day
new_sales
```

```python
new_sales.drop('Date', axis=1, inplace=True)
```

```python
new_sales.dtypes
```

```python
new_sales['Day_of_week'].value_counts()
```

```python
new_sales.drop('Open', axis=1, inplace=True)
```

```python
new_sales
```

```python
fig = plt.figure(figsize=[12,5])
sns.heatmap(new_sales.corr(), annot=True)
plt.show()
```

```python
features = new_sales.loc[:, ~new_sales.columns.isin(['Sales'])]
label = new_sales['Sales']
```
